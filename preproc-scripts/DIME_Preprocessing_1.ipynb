# Reducing S_ESTS to only candidates with recipient info in DIME

```python
dfs = pd.read_excel('S_ESTS.xlsx',header=None,
                    names=['CNUM','ICPSR','STATE','DISTRICT','STATENAME','PARTY','NAME','DIM1','DIM2','1','2','3','4','5','6','7'])
dfs2 = pd.read_excel('H_ESTS.xlsx',header=None,
                    names=['CNUM','ICPSR','STATE','DISTRICT','STATENAME','PARTY','NAME','DIM1','DIM2','1','2','3','4','5','6','7'])
dfs = dfs.append(dfs2)
dfr = pd.read_csv('DIME_RECIP.csv')
dfs = dfs[dfs['CNUM'] >= 110]
dfr = dfr[np.logical_and(dfr['cycle'] >= 2006,dfr['cycle'] <= 2012)]
# dfr = dfr[~dfr['ICPSR'].str.contains('nominee')]
temp = dfr['ICPSR2'].values
temp2 = [t[-5:] for t in temp]
temp2 = [re.sub("\D","",s) for s in temp2]
# temp2 = [t.lstrip('0') for t in temp2]
dfr['ICPSR2'] = temp2
dfsdict = dfs['ICPSR'].to_dict()
print(dfr.shape)
dfr = dfr.reset_index()
dfnew = pd.DataFrame(columns=dfr.columns)
dfTemp = pd.DataFrame(columns=['RID'])
count = 0
misParse = 0
for i, row in dfr.iterrows():
    try:
        if int(row['ICPSR2']) in dfsdict:
            dfnew = dfnew.append(row)
            count += 1
    except Exception as e:
        misParse += 1
        print(e)
    if i%1000 == 0:
        print('I:' + str(i))
    if count%1000 == 0:
        print('count:' + str(count))
    if misParse % 1000 == 99:
        print('parse:' + str(misParse))
print(dfnew.shape)
```
